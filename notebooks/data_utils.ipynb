```python
from src.data_utils import (
    load_data,
    clean_numeric_data,
    balance_classes,
    split_and_scale
)
```

## Preprocessing Workflow

The dataset is prepared for model training by first cleaning numeric columns, then splitting into train and test sets, and finally balancing the training set using SMOTETomek.

```python
# 2. Split and scale first
X_train, X_test, y_train, y_test, scaler = split_and_scale(
    df,
    target_column="status"
)


X_train_balanced, y_train_balanced = balance_classes(
    X_train,
    y_train,
    method="tomek"
)

print("Cleaned data shape:", df.shape)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Balanced train shape:", X_train_balanced.shape)
print("Balanced target counts:")
print(y_train_balanced.value_counts())
```
